L'idée initale était de récuperait les 500 entreprises du S&P 500, qui est un indice boursier avec les 500 entreprises américaines ayant la plus grande capitalisation boursière.
Cependant, la quantité de données étant trop conséquent, j'ai decidé de prendre uniquement les 10 premières.

```python
import yfinance as yf
import pandas as pd
import requests

# Step 1: Récupérer la liste des symboles des actions
#url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
#tables = pd.read_html(url)
#sp500_df = tables[0] # La première table est celle qui nous intéresse
#tickers = sp500_df["Symbol"].tolist()
#tickers = ["AAPL", "NVDA", "MSFT", "AMZN", "GOOGL","META","TSLA","BRYN.HA","UNH","JNJ","V","PG","JPM","HD","MA","CVX","PFE","XOM","ABBV","PEP"]
tickers = ["AAPL", "NVDA", "MSFT", "AMZN", "GOOGL","META","TSLA","BRYN.HA","UNH","JNJ"]
# Step 2: télécharger les données historiques pour chaque action
data_list = []
for ticker in tickers:
    try:
        stock_data = yf.download(ticker, start="2015-01-01", end="2024-12-31", interval="1d")
        stock_data["Ticker"] = ticker  # On ajoute une colonne pour le symbole de l'action
        stock_data.reset_index(inplace=True)
        stock_data.columns = [None] * len(stock_data.columns)
        data_list.append(stock_data)
    except:
        print(f"Failed to get data for {ticker}")

# Step 3: Combiner les données en un seul DataFrame
all_stocks_df = pd.concat(data_list, ignore_index=True)
```

On récupere le df, et on prend uniquement les colonnes qui nous intéresse.

```python
all_stocks_df.columns = ["Date", "Close", "High", "Low", "Open", "Volume", "Ticker"]
all_stocks_df
```

On exporte les données sous forme de fichier Excel.

```python
all_stocks_df.to_excel('data.xlsx')
```
